```python
import string
from string import maketrans
import pandas as pd
```

```python
dflist = []

dflist.append(pd.read_csv('Youtube01-Psy.csv'))
dflist.append(pd.read_csv('Youtube02-KatyPerry.csv'))
dflist.append(pd.read_csv('Youtube03-LMFAO.csv'))
dflist.append(pd.read_csv('Youtube04-Eminem.csv'))
dflist.append(pd.read_csv('Youtube05-Shakira.csv'))

concatdf = pd.concat(dflist,axis = 0)
```

```python
del concatdf['COMMENT_ID']
del concatdf['AUTHOR']
del concatdf['DATE']
concatdf
```

```python
spamdict = dict() #dictionary with all spam words in dataset
hamdict = dict() #dictionary with all ham words in dataset
count = 0
word = ""

slinecount = 0
hlinecount = 0
numlines = 1368
pA = 0 #probability that a word is spam
pnotA = 0 #probability that a word is not spam



for j in range(1368):
    if concatdf.iloc[j, 1] == 1:
        slinecount += 1
    else:
        hlinecount += 1
        

pA = (float(slinecount) / numlines)
pnotA = (float(hlinecount) / numlines)

    
#code to extract words from datset and put it into the right dictionary
for i in range(1368):
    comment = concatdf.iloc[i,0]
    comment = comment.lower()
    comment = ''.join(i for i in comment if not i.isdigit())
    for letter in comment:
        if letter == ' ' or count == len(comment):
            if word != ' ' and word != '':
                if concatdf.iloc[i,1] == 1:
                    if word not in spamdict:
                        spamdict[word] = 1
                    else:
                        spamdict[word] += 1
                else:
                    if word not in hamdict:
                        hamdict[word] = 1
                    else:
                        hamdict[word] += 1
                word = ""                
        else:
            word += letter
                
        count += 1        

#takes a given string -sampleTest- and classifies as spam or ham
def classify(hamdict, spamdict, sampleTest):
    hamsum = 0
    spamsum = 0

    for key,value in hamdict.items():
        hamsum += value

    for key,value in spamdict.items():
        spamsum += value

    pwordgivenSpam = 0
    pwordgivenHam = 0

    word2 = ""
    stringcount = 0

    for letter in sampleTest:
        if letter == ' ':
            word2 = word2.strip()
            if word2 in spamdict:
                pwordgivenSpam = pwordgivenSpam + (spamdict[word2]/float(spamsum))

            if word2 in hamdict:
                pwordgivenHam = pwordgivenHam + (hamdict[word2] / float(hamsum))

            word2 = ""

        if stringcount == ((len(sampleTest)) - 1):
            word2 += letter
            word2 = word2.strip()
            if word2 in spamdict:
                pwordgivenSpam = pwordgivenSpam + (spamdict[word2] / float(spamsum))

            if word2 in hamdict:
                pwordgivenHam = pwordgivenHam + (hamdict[word2] / float(hamsum))

            word2 = ""

        else:
            word2 += letter
        stringcount += 1

    if (pwordgivenSpam * pA) > (pwordgivenHam * pnotA):
        return 1
    else:
        return 0
    
#------------------------------------------------------------    

total = 585
correct = 0
for x in range (1369,1954):
    num = classify(hamdict, spamdict, concatdf.iloc[x,0])
    if concatdf.iloc[x,1] == 1:
        if num == 1:
            correct +=1
    else:
        if num == 0:
            correct += 1
            
accuracy = 0
accuracy = (correct / float(total))
accuracy = accuracy*100
print("This is the accuracy of calculating whether it is spam or ham: ")
print accuracy
SPAM = classify(hamdict, spamdict, "This video is good")

if(SPAM==1):
    print("The comment is SPAM")
else:
    print("The comment is HAM")
    
```
